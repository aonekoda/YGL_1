# 함수 (Function)

```python
#함수를 선언한다. 
def sum(num1, num2) :
    result = num1+num2
    return result

# 함수를 호출한다.

num = sum(100, 200)
print(num)
```

### 기본값이 있는 매개변수(parameter)

```python
def printInfo (name, job, age=20, address='서울') :
    print('이름:{}, 직업:{}, 나이:{}, 주소:{}'.format(name,job,age,address))
    
printInfo(name='이선화',job='강사')
printInfo(address='인천', name='강동원', job='배우')
```

### 가변인자의 사용

```python
def print_func(* num) :
    for i in num:
        print(i)

print_func('이선화','정우성','심은하', '박보검','강동원')
```

```python
def printinfo(** info) :
    for a in info.keys() :
        print(info[a])

printinfo(name='이선화', job='강사', age=20)
```

# 함수 실습

두 개의 숫자를 입력으로 받아 합을 리턴하는 함수를 만든다.

```python
def my_sum(num1, num2):
    return num1+num2

print(my_sum(90, 250))
```

가변인자를 입력으로 받아 숫자들의 평균을 리턴하는 함수를 생성한다.

```python
def avg_score(* score):
    tot= 0
    for i in score:
        tot = tot + i
    avg = tot / len(score)
    return avg

print(avg_score(90,80,43,84.100))
```
